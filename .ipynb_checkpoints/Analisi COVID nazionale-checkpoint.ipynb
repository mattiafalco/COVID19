# Analisi COVID19 andamento nazionale

Questo notebook è pensato per chi vuole essere aggiornato sull'andamento dei contagi della malattia COVID19 in Italia. I dati utilizzati sono raccolti dal repository https://github.com/pcm-dpc.

## Caricamento dati 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from matplotlib import rcParams
import seaborn as sns

sns.set()

# Arial, default font
rcParams['font.sans-serif'] = "Arial"
rcParams['font.family'] = "sans-serif"

years = mdates.YearLocator()   # every year
months = mdates.MonthLocator()  # every month
days = mdates.DayLocator(interval=2) # every day
years_fmt = mdates.DateFormatter('%d-%m-%Y')
```

```python
url = 'https://raw.githubusercontent.com/pcm-dpc/COVID-19/master/dati-andamento-nazionale/dpc-covid19-ita-andamento-nazionale.csv'
df = pd.read_csv(url)
print( 'Dati aggiornati al giorno ', df['data'].iloc[-1])
df.head()
```

```python
# colonna giorno
df['giorno'] = df['data'].str.slice(start=0, stop=10)

primo_giorno = df['giorno'].iloc[0]
ultimo_giorno = df['giorno'].iloc[-1]

giorni = pd.date_range(start=primo_giorno, end=ultimo_giorno, freq='D')
df['giorno']=giorni
```

```python
# colonna nuovi tamponi
df['nuovi_tamponi'] = df['tamponi'].diff()
# colonna nuovi guariti
df['nuovi_guariti'] = df['dimessi_guariti'].diff()
# colonna nuovi deceduti
df['nuovi_deceduti'] = df['deceduti'].diff()

df.to_csv('dati_andamento_nazionale')
df
```

```python
# funzione grafici
def Set_graphix(ax, legend=True, grid=True):
    # format the ticks
    ax.xaxis.set_major_locator(months)
    ax.xaxis.set_major_formatter(years_fmt)
    ax.xaxis.set_minor_locator(days)
    
    plt.sca(ax)
    plt.xticks(fontsize='large')
    plt.yticks(fontsize='large')
    #sns.set_style("ticks", {"xtick.major.length": 8, "xtick.major.width":4})
    plt.rcParams['xtick.bottom'] = True
    plt.rcParams['xtick.major.size'] = 4
    plt.rcParams['xtick.major.width'] = 4

    # round to nearest days.
    datemin = np.datetime64(primo_giorno, 'D')-1
    datemax = np.datetime64(ultimo_giorno, 'D') +5
    ax.set_xlim(datemin, datemax)
    
    ax.grid(grid)

    # rotates and right aligns the x labels, and moves the bottom of the
    # axes up to make room for them
    fig.autofmt_xdate(ha='center')
 
    #set legend
    if legend:
        ax.legend(fontsize='x-large', loc='upper left' )
```

## Andamento contagi

Confrontiamo l'andamento dei contagi con guariti e decessi.

```python
# dati
positivi = df['totale_positivi']
guariti = df['dimessi_guariti']
deceduti = df['deceduti']
#sns.set_style('white')
fig, ax = plt.subplots(figsize=(8,6))
ax.plot(giorni, positivi,'#0d47a1', label='positivi', linewidth=3)
ax.plot(giorni, guariti, '#039be5', label='guariti', linewidth=3)
ax.plot(giorni, deceduti, '#e65100', label='deceduti', linewidth=3)
ax.set_title('Contagi', fontsize=24, pad=20, fontweight = 'bold' )

Set_graphix(ax, grid=False)
sns.despine()

plt.show()
```

## Confronto ricoverati-isolamento

```python
#dati
ricoverati = df['ricoverati_con_sintomi']
ric_gravi = df['terapia_intensiva']
isolamento = df['isolamento_domiciliare']
ric_totali = df['totale_ospedalizzati']

#plot
fig, ax = plt.subplots(figsize=(8,6))
ax.plot(giorni, ricoverati, 'b', label='ricoverati')
ax.plot(giorni, ric_gravi, 'r', label='terapia intensiva')
ax.plot(giorni, isolamento, 'g', label='isolamento')
ax.plot(giorni, positivi, 'black', label='totale_positivi')
ax.set_title('Confronto ricoverati-isolamento', fontsize=24, pad=20, fontweight='bold' )
Set_graphix(ax)
plt.show()

print('Dati attuali' )
print( 'Positivi: {}'.format(positivi[len(positivi)-1]))
print( 'Isolamento: {}'.format(isolamento[len(isolamento)-1]))
print( 'Ricoverati: {}'.format(ricoverati[len(ricoverati)-1]))
print( 'Terapia intensiva: {}'.format(ric_gravi[len(ric_gravi)-1]))
```

```python
#plot
fig, ax = plt.subplots(figsize=(8,6))
colors = ['#0d47a1','#039be5', '#e65100']
labels = ['Ricoverati', 'Terapia intensiva', 'Isolamento']
sns.set_style('white')
ax.stackplot(giorni,
             isolamento,
             ricoverati,
             ric_gravi,
             colors = colors,
             edgecolor = 'none')
Set_graphix(ax, grid=False, legend=False)
ax.legend(labels = labels, loc = 'upper left')
ax.set_title('Confronto ricoverati-isolamento', fontsize=24, pad=20, fontweight='bold' )
sns.despine()
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8,6))
ax.plot(giorni, ric_totali, label='totale ospedalizzati', color=colors[0])
ax.plot(giorni, isolamento, color=colors[1], label='isolamento')
ax.set_title('Confronto ricoverati-isolamento', fontsize=24, pad=20, fontweight='bold' )
Set_graphix(ax, grid=False)
sns.despine()
plt.show()
```

```python
#plot
fig, ax = plt.subplots(figsize=(8,6))
colors = ['#0d47a1','#039be5', '#e65100']
labels = ['Ricoverati', 'Terapia intensiva', 'Isolamento']
sns.set_style('white')
ax.stackplot(giorni,
             isolamento,
             ric_totali,
             colors = colors,
             edgecolor = 'none')
Set_graphix(ax, grid=False, legend=False)
ax.legend(labels = labels, loc = 'upper left')
ax.set_title('Confronto ricoverati-terapia_intensiva', fontsize=24, pad=20, fontweight='bold' )
sns.despine()
plt.show()
```

```python
#dati
frazione_ric = ric_totali/isolamento
#plot
fig, ax = plt.subplots(figsize=(8,6))
ax.plot(giorni, frazione_ric, 'navy', label='rapporto ricoverati/isolamento')
ax.set_title('Rapporto ricoverati/isolamento', fontsize=24 , pad=20, fontweight='bold')
Set_graphix(ax, grid=False)
sns.despine()
plt.show()
```

## Confronto tamponi-positivi

```python
#dati
tamponi = df['tamponi']
totale_casi = df['totale_casi']
fraz1 = totale_casi/tamponi*100

casi_testati = df['casi_testati']
fraz2 = totale_casi/casi_testati*100

#plot
fig, ax = plt.subplots(figsize=(8,6))
ax.plot(giorni, fraz1, color=colors[0], label='percentuale positivi/tamponi', )
ax.plot(giorni, fraz2, color=colors[1], label='precentuale positivi/casi_testati')
ax.set_title('Confronto tamponi-positivi', fontsize=24, pad=20, fontweight='bold' )
Set_graphix(ax, grid=False)
ax.legend(loc='upper right', fontsize='x-large')
sns.despine()
plt.show()


print( 'Frazione tamponi positivi: ', fraz1[len(fraz1)-1],'%')
print( 'Frazione casi positivi: ', fraz2[len(fraz2)-1],'%')
```

## Confronto aumento tamponi/positivi

```python
#dati
nuovi_positivi = df['nuovi_positivi']

aumento_tamponi = df['tamponi'].diff()
fraz1 = nuovi_positivi/aumento_tamponi*100

aumento_casi_testati = df['casi_testati'].diff()
fraz2 = nuovi_positivi/aumento_casi_testati*100

#plot
fig, ax = plt.subplots(figsize=(8,6))
ax.plot(giorni, fraz1, color=colors[0], label='positivi/tamponi %')
ax.plot(giorni, fraz2, color=colors[1], label='positivi/casi_testati %')
ax.set_title('Confronto aumento tamponi-positivi', fontsize=24, pad=20 ,fontweight='bold')
Set_graphix(ax, grid=False)
ax.legend(loc='upper right', fontsize='x-large')
sns.despine()
plt.show()


print( 'Frazione tamponi positivi: {}%'.format(np.around(fraz1[len(fraz1)-1], 2)))
print( 'Frazione casi positivi: {}%'.format(np.around(fraz2[len(fraz2)-1], 2)))
```

## Variazione e nuovi positivi

```python
#dati
variazione = df['variazione_totale_positivi']

#plot
fig, (ax1, ax2) = plt.subplots(1,2, figsize=(20,10))
ax1.plot(giorni, nuovi_positivi, 'navy')
ax1.set_title('Nuovi positivi', fontsize=24, pad=20, fontweight='bold')

Set_graphix(ax1, legend=False)

ax2.plot(giorni, variazione, 'navy')
ax2.set_title('Variazione', fontsize=24, pad=20, fontweight='bold')
Set_graphix(ax2, legend=False)


plt.show()
```
